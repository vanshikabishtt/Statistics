```python
import pandas as pd
import numpy as np
```

```python
data={
    "gender":["Male","Male","Female","Male","Female","Female","Male","Female","Male","Female"],
      "result": ['pass','fail','fail','pass','fail','fail','fail','fail','pass','fail']
}

df=pd.DataFrame(data)
print(df)
```

```python
table=pd.crosstab(df["gender"],df["result"])
table
```

```python
from scipy.stats import chi2_contingency
chi2,p,dof,expectancy=chi2_contingency(table)
print("Value of Chi2 is:",chi2)
print("Value of p is:",p)
print("Value of dof :",dof)
print("Value of expectancy",expectancy)

if p>0.05:
  print("There is association between gender and result(Fail to reject H0)")
else:
  print("There is no association between gender and result(reject H0)")
```
