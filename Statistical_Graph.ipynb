# **Bar Graph**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
categories=["a","b","c","d","e"]
values=[5,7,3,8,4]
plt.bar(categories,values,color='lightcoral',edgecolor='teal')
plt.xlabel("Student Name")
plt.ylabel("Marks in Stats")
plt.title("Example Bar Chart")
plt.show()
```

```python
categories=["a","b","c","d","e"]
values=[5,7,3,8,4]
plt.bar(categories,values,color='palevioletred',edgecolor='maroon')
plt.xlabel("Student Name")
plt.ylabel("Marks in Stats")
plt.title("Mid Term Marks")
plt.grid(axis='y',linestyle="--",alpha=0.7)
plt.show()
```

# **Line Graph**

```python
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
plt.plot(x,y,color="darkcyan",linestyle="--",marker="o",label="y=x*2")
plt.legend()
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Chart Example")
plt.show()
```

# **Scatter Plot**

```python
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
sizes=[10,20,30,40,50,60]
plt.scatter(x,y,color="mediumvioletred",s=sizes,edgecolor="teal",marker="o",label="y=x*2")
plt.legend()
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Chart Example")
plt.show()
```

# **Histogram**

```python
np.random.seed(30) # For Fixing
data=np.random.randn(1000)
plt.hist(data,bins=20,color="powderblue",alpha=0.7,edgecolor="black",label="datapoint")
plt.grid()
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.show()
```

# **PieChart**

```python
sizes=[40,30,20,10]
labels=["Group 1","Group 2","Group 3","Group 4"]
colors=["teal","crimson","violet","mediumvioletred"]
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=140,colors=colors,explode=(0.1,0,0,0))
plt.title("Pie Chart example")
plt.show()
```

```python
categories=["a","b","c","d","e"]
values=[5,7,3,8,4]
widths=[0.1,0.5,0.7,0.2,0.3]
colors=["teal","cyan","indianred","violet","crimson"]
plt.bar(categories,values,width=widths,color= colors,edgecolor='teal')
plt.xlabel("Student Name")
plt.ylabel("Marks in Stats")
plt.title("Example Bar Chart")
plt.show()
```
